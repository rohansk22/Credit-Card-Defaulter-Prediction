Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=0.2,random_state=3)
```

```python
y.value_counts()
```

The data set contains an imbalance. There are significantly higher number of 'N' or No than 'Y' or Yes. We will try to balance the two by importing more 'Y' data into the data set. This is called as oversampling the imbalance. Similarly if we remove records, it is called undersampling. Here, we will balance the data by oversampling it using the SMOTE library.

```python
#Balancing records
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=3)
```

```python
x_balanced,y_balanced = smote.fit_resample(x_final,y)
```

```python
y_balanced.value_counts()
```

```python
train_x,test_x,train_y,test_y = train_test_split(x_balanced,y_balanced,test_size=0.2,random_state=3)
```

```python
model.fit(train_x,train_y)
```

```python
pred_y= model.predict(test_x)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_y,pred_y))
```

The F1 score isn't that high. Ideally, an F1 score of 0.8 or greater is a good fit. 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier()
```

```python
model_RF.fit(train_x,train_y)
```

```python
pred_y_RF = model_RF.predict(test_x)
```

```python
print(classification_report(test_y,pred_y_RF))
```

```python
train_y.value_counts()
```

The Random Forest model predicted the defaulters with an F1 score of 0.85 and an accuracy of 0.85. This is a more powerful model and hence fits the train data much better.
